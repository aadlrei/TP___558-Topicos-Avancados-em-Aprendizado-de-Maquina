```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Rutas en tu Drive
audio1 = "/content/drive/MyDrive/Audios/waveNet/normal_1.wav"
audio2 = "/content/drive/MyDrive/Audios/waveNet/audioVEED2.mp3"
audio3 = "/content/drive/MyDrive/Audios/waveNet/audioVEED1.mp3"

# Cargar audios
y1, sr1 = librosa.load(audio1, sr=None)
y2, sr2 = librosa.load(audio2, sr=None)
y3, sr3 = librosa.load(audio3, sr=None)

# Crear figura
plt.figure(figsize=(12, 6))

# Waveform 1
plt.subplot(3, 1, 1)
librosa.display.waveshow(y1, sr=sr1, alpha=0.7)
plt.title("Waveform - Audio 1")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")

# Waveform 2
plt.subplot(3, 1, 2)
librosa.display.waveshow(y2, sr=sr2, alpha=0.7, color="orange")
plt.title("Waveform - Audio 2")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")

# Waveform 3
plt.subplot(3, 1, 3)
librosa.display.waveshow(y3, sr=sr3, alpha=0.7, color="green")
plt.title("Waveform - Audio 3")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y2, sr=sr2, alpha=0.7, color="orange", label="Audio 2")
librosa.display.waveshow(y3, sr=sr3, alpha=0.7, color="green", label="Audio 3")
librosa.display.waveshow(y1, sr=sr1, alpha=0.7, color="blue", label="Audio 1")
plt.title("Comparacin de Waveforms")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.show()
```
